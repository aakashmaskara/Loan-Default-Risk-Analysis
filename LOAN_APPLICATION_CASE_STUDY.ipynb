#### AMT_ANNUITY
- Since the percentage of null value is very small (<1%), the data must be skewed. So we can replace the missing values with Median

```python
df1.AMT_ANNUITY.fillna(df1.AMT_ANNUITY.median(), inplace = True)
```

#### AMT_GOODS_PRICE
- Since the percentage of null value is very small (<1%), the data must be skewed. So we can replace the missing values with Median

```python
df1.AMT_GOODS_PRICE.fillna(df1.AMT_GOODS_PRICE.median(), inplace = True)
```

#### NAME_TYPE_SUITE
- Since the percentage of null value is very small (<1%), null values can be replaced with Mode of the dataset, since the column is categorical

```python
df1.NAME_TYPE_SUITE.value_counts()
```

```python
df1.NAME_TYPE_SUITE.fillna(df1.NAME_TYPE_SUITE.mode()[0], inplace=True)
```

#### OCCUPATION_TYPE
- Since the null value percentage is very high and this column is a critical part of the analysis, replacing the null values with mode might not be the ideal way to deal with this situation. It would be preferred to label these null values as Unknown

```python
df1.OCCUPATION_TYPE.fillna("Unknown", inplace=True)
```

#### EXT_SOURCE_2
- Since the percentage of null value is very small (<1%), null values can be replaced with Mode of the dataset, since the column is categorical

```python
df1.EXT_SOURCE_2.fillna(df1.EXT_SOURCE_2.median(), inplace=True)
```

## Dealing with columns which have Null Values

#### AMT_ANNUITY

```python
# Since Loan Annuity is a specific amount deducted every month we would first need to check if the loan has been approved, rejected
# or refused. If there isnt any case of loan approved then we can simply mark that row as 0 for AMT_ANNUITY
df2.NAME_CONTRACT_STATUS.value_counts()
```

```python
# We can assign a value of 0 for the rows which have their loans rejected or refused.
rejected_rows = ['Canceled', 'Refused', 'Unused offer']
df2.loc[df2.NAME_CONTRACT_STATUS.isin(rejected_rows),'AMT_ANNUITY']=0
```

```python
df2.AMT_ANNUITY.isnull().mean()*100
```

```python
# There are still a few null values and since the value is very small we can replace them with Median
df2.AMT_ANNUITY.fillna(df2.AMT_ANNUITY.median(), inplace=True)
```

#### AMT_GOODS_PRICE

```python
df2.AMT_GOODS_PRICE.plot.hist(bins=500)
```

```python
# Since the data seems skewed we can replace all Null values with median.
df2.AMT_GOODS_PRICE.fillna(df2.AMT_GOODS_PRICE.median(), inplace=True)
```

```python
# One sanity check for us could be that the loans which have been approved should have AMT_GOODS_PRICE less than or equal to AMT_CREDIT
# Dropping the rows where the condition is not met.
df2.loc[~(df2.AMT_GOODS_PRICE > df2.AMT_CREDIT) & (df2.NAME_CONTRACT_STATUS == 'Approved')]
```

#### CNT_PAYMENT

```python
df2.CNT_PAYMENT.plot.hist(bins=500)
```

```python
# The data does not seem to be skewed, so we can replace all null values with Mean.
df2.CNT_PAYMENT.fillna(df2.CNT_PAYMENT.mean(), inplace=True)
```

#### PRODUCT_COMBINATION

```python
# Since PRODUCT_COMBINATION has object data types. We can replace the null values with mode
df2.PRODUCT_COMBINATION.fillna(df2.PRODUCT_COMBINATION.mode()[0], inplace=True)
```
